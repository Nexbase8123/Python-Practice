```python
# Lets practice!
# ................lists and how to manipulate lists
# To create a list, used the square brackets [] to enclose the items or elements or values of the list separated by commas.

# Example; Seyi was tolde to withdraw 5000 from pos to buy rice, beans, egg, meat, spaggeti and salad. Create a list snd all of the items in the list
```

```python
seyi_list = ["rice", "beans", "egg", "meat", "spag", "salad"]
print(seyi_list)
```

```python
# Example 2; Add the amount she withrew to the list. Hint: Use the dot append() function
# how to use append
# your_list.append(item)

seyi_list.append(5000)
print(seyi_list)
```

```python
# Example 3; Seyi was later told to by table_water when coming, insert table water before rice in the list. Hint: use the insert function
# how to use insert
# your_list.insert(index, item) # index means the position you would want it to be starting from 0 to n(last number in your list)
# So the first index in your list is zero(0)

seyi_list.insert(0, "table_water")
print(seyi_list)
```

```python
# Example 4; Seyi was again told to return the egg and buy kpomo in place of egg. Hint: Use the dot remove() function
# how to use remove
# your_list.remove(item)

seyi_list.remove("egg")
print(seyi_list)
```

```python
# Add the kpomo in the place of egg. Hint: Use insert()function
seyi_list.insert(3, "kpomo") ####
print(seyi_list)
```

```python
# Example 5; We notice that the spelling of spagetti was shortened, lets correct it.
# how to modify an item in a list.
# your_list[index_of_the_item_you_want_to_replace] = "new_item"

seyi_list[5] = "spaggetti"
print(seyi_list)
```

```python
#Example 6; There is no change remaining, hence you should remove the money from the list
# your_list.remove(item)
seyi_list.remove(5000)
print(seyi_list)
```
