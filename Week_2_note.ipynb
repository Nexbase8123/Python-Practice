```python
#...............Dictionary and how to manipulate  dictionaries.............

""" A Python dictionary is a collection of key-value pairs. You can create a dictionary using curly braces {}.
You can access the values using their keys"""
```

```python
# lets create a dictionary
joy_dictionary = {"name": "Joy", "age": 25, "city": "Lagos"}
joy_dictionary
```

```python
# lets use the keys to access the values
joy_dictionary["name"]
```

```python
# lets get the age
joy_dictionary["age"]
```

```python
#......................iterating through a dictionary...................

#lets get all the keys
for key in joy_dictionary:
  print(key)
```

```python
# lets get all the values. Hint: use the .value() function
for value in joy_dictionary.values():
  print(value)
```

```python
# lets get both keys and values. Hint: use the .items() function
for key, value in joy_dictionary.items():
  print(key, value)
```

```python
# lets add joy's education background to the dictionary
joy_dictionary["education"] = "BSc"
print(joy_dictionary)
```

```python
 # lets reduce joys' age
joy_dictionary["age"] -= 1
print(joy_dictionary)
```

```python
# lets add two to her age
joy_dictionary["age"] += 2
print(joy_dictionary)
```

```python
# lets replace her age with 28
joy_dictionary["age"] = 28
print(joy_dictionary)
```

```python
# lets remove her age. Hint: use 'del'
del joy_dictionary["age"]
print(joy_dictionary)
```

```python
# lets check if her age is still there
"age" in joy_dictionary
```

```python
# lets check using another method(have a little task of conditional statement)
if "age" in joy_dictionary:
  print("Age is still there")
else:
  print("Age is not there")
```

```python
## NOTE 1. To get only keys use .key() 2. To get only values use.values() 3. To get both use .items()
## 4. To combine two dictionaries, use .update() 5. To clear everything in a dictionary use .clear
```
